```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
from scipy import stats
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import OneHotEncoder
import statsmodels.api as sm
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
#1 - Get the numerical variables from our dataset.
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head(3)
```

```python
data.dtypes
```

```python
numerical = data.select_dtypes(include=[int, float])
numerical.head()
```

```python
# 2 - Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

#some of the variables represent theoretical normal such as 'distribution customer lifetime value' or 'distribution of total claim amount'
```

```python
for col in numerical:
    sns.distplot(numerical[col], kde=True, hist=False)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```

```python
# 3 - #Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
```

```python
numerical_log = numeric.apply(lambda x: np.log(x))
numerical_log.head()
```

```python
for col in numerical_log:
    sns.distplot(numerical_log[col], kde=True, hist=False)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```
